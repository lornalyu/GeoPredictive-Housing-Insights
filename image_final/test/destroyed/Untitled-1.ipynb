```python


# But there is not output, so we need to print the output
import cv2
import os
import sys

def resize(input_dir, output_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.endswith('.tif'):
                img = cv2.imread(os
                                    .path.join(root, file), cv2.IMREAD_UNCHANGED)
                img = cv2.resize(img, (2024, 2024))
                cv2.imwrite(os.path.join(output_dir, file), img)
                print(f'Processed {file}')

if __name__ == '__main__':
    if len(sys.argv) != 3:
        print('Usage: python3 resize.py <input_dir> <output_dir>')
    else:
        input_dir = sys.argv[1]
        output_dir = sys.argv[2]
        resize(input_dir, output_dir)

    print('Done')

# Now run the code with the input and output directories
input_dir = '/Users/aolyu/Documents/GitHub/QTM-Project/image_final'
output_dir = '/Users/aolyu/Documents/GitHub/QTM-Project/image_final1'
resize(input_dir, output_dir)
```
